# Dataframe

```python
import pandas as pd

df = pd.read_excel("board2.xlsx")
df['Average'] = df.mean(numeric_only=True, axis=1)
df
```

```python
print("--Instru--")
Instru = df.loc[(df.Track=="Instrumentation") & (df.Hometown=="Luzon") & (df.Electronics>70) ,['Name','GEAS','Electronics']]
Instru
```

```python
print("--Mindy--")
Mindy = df.loc[(df.Hometown=="Mindanao") & (df.Gender=="Female") & (df.Average>=55) ,['Name','Track','Electronics', 'Average']]
Mindy
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,6))
plt.bar(df['Track'],df['Average'])
plt.xlabel("Chosen Track in College")
plt.ylabel("Average")
plt.title("Table 1.1: A bar graph showing the relationship of chosen track in college and average grade.")
```

```python
plt.figure(figsize=(5,6))
plt.bar(df['Gender'],df['Average'])
plt.xlabel("Gender")
plt.ylabel("Average")
plt.title("Table 1.2: A bar graph showing the relationship of gender and average grade.")
```

```python
plt.figure(figsize=(5,6))
plt.bar(df['Hometown'],df['Average'])
plt.xlabel("Hometown")
plt.ylabel("Average")
plt.title("Table 1.3: A bar graph showing the relationship of hometown and average grade.")
```

```python
#Based on the three graphs above, the chosen track, gender, and the hometown affects the average grade.
#Microelectronics in the chosen track shows better average grade.
#Females have higher grades relative to males.
#People who have Luzon as their hometown have better average grades than the rest.
```
